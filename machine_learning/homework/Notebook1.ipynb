## Exercise #6

Do something cool. Using the movies dataset, get creative and show me something about the dataset. Use a text cell to explain what you did.

##Explanation
Using a double nested loop, the below program runs through each unique year that movies were published. For each year, it runs through each movie in the dataset. For the unique year it is looking at, it counts the number of movies of each genre. When the inner loop gets to a year that is beyond the unique year, the inner loop is broken. When the program finishes, you are left with a large printed list of years and the genre of movie that was most common in that year. With a large dataset, of course there are some years where the max genre is the same for mutiple genres.

```python
import numpy as np
movies_by_year = movies.sort_values(by=['Year'])
movies_by_year = movies_by_year.reset_index(drop=True)
list_of_genres = movies['Genre'].unique()

for i in movies_by_year['Year'].unique():
  genre_count = np.zeros(list_of_genres.size)
  for x in range (movies_by_year['Year'].size):
    if (movies_by_year['Year'].iloc[x] > i):
      break
    #print(movies_by_year['Genre'].iloc[x])
    if (movies_by_year['Year'].iloc[x] == i):
      genre_count[np.where(list_of_genres == movies_by_year['Genre'].iloc[x])] += 1
  
  highest_genre = genre_count.max()
  print('Year:', i)
  print('Most movies produced in a genre:', highest_genre, list_of_genres[np.where(genre_count == highest_genre)])
  print()
```
