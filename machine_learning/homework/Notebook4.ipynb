```python
# import libraries
import sklearn
import pandas
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import neighbors

# load data
housing_data = pandas.read_csv('/content/drive/MyDrive/CS167 Datasets/HousingData.csv')
housing_data
```

# Part 1: Clean and normalize the Data
Clean and normalize the data (use the StandardScalar from sklearn). Replace any null values with the average of the columns. Verify that there are no null values by printing out the results of `housing_data.isna().any()`

```python
# clean data here!!!
housing_data['CRIM'].fillna(housing_data['CRIM'].mean(), inplace=True)
housing_data['ZN'].fillna(housing_data['ZN'].mean(), inplace=True)
housing_data['INDUS'].fillna(housing_data['CRIM'].mean(), inplace=True)
housing_data['CHAS'].fillna(housing_data['CHAS'].mean(), inplace=True)
housing_data['AGE'].fillna(housing_data['AGE'].mean(), inplace=True)
housing_data['LSTAT'].fillna(housing_data['LSTAT'].mean(), inplace=True)
housing_data.isna().any() # look at all of these columns with null values!

#normalization will be in a later part after we split the data
```

```python
# Split the data into the training data and testing data
target= 'MEDV'
predictors = housing_data.columns.drop(target) # use all of the columns except for MEDV
train_data, test_data, train_sln, test_sln = train_test_split(housing_data[predictors], housing_data[target], test_size = 0.2, random_state=41)
train_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

#normalization
scaler = StandardScaler()
scaler.fit(train_data)
train_data_normalized = scaler.transform(train_data)
test_data_normalized = scaler.transform(test_data)
```

```python
# kNN with small k
small_knn = neighbors.KNeighborsRegressor(n_neighbors=5)
small_knn.fit(train_data_normalized, train_sln)
small_k_predict = small_knn.predict(test_data_normalized)
print('r2:', metrics.r2_score(test_sln, small_k_predict))
```

```python
# kNN with large k
large_knn = neighbors.KNeighborsRegressor(n_neighbors=25)
large_knn.fit(train_data_normalized, train_sln)
large_k_predict = large_knn.predict(test_data_normalized)
print('r2:', metrics.r2_score(test_sln, large_k_predict))
```

```python
# w-knn with small k
w_small_knn = neighbors.KNeighborsRegressor(n_neighbors=5, weights='distance')
w_small_knn.fit(train_data_normalized, train_sln)
w_small_k_predict = w_small_knn.predict(test_data_normalized)
print('r2:', metrics.r2_score(test_sln, w_small_k_predict))
```

```python
# w-knn with large k
w_large_knn = neighbors.KNeighborsRegressor(n_neighbors=25, weights='distance')
w_large_knn.fit(train_data_normalized, train_sln)
w_large_k_predict = large_knn.predict(test_data_normalized)
print('r2:', metrics.r2_score(test_sln, w_large_k_predict))
```

```python
# decision tree with default paramters
dt = tree.DecisionTreeRegressor(random_state=0)
dt.fit(train_data, train_sln)
dt_predict = dt.predict(test_data)
print('r2:', metrics.r2_score(test_sln, dt_predict))
```

```python
# decision tree with some kind of parameter that (hopefully) results in a smaller tree
small_dt = tree.DecisionTreeRegressor(random_state=0, max_depth=4)
small_dt.fit(train_data, train_sln)
small_dt_predict = dt.predict(test_data)
print('r2:', metrics.r2_score(test_sln, small_dt_predict))
```
