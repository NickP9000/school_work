```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
data= pd.read_csv('/content/drive/MyDrive/CS167 Datasets/breast-cancer-wisconsin-data.csv')
data.head()
```

```python
#we need to drop the last column 'Unnamed: 32'
data.drop(labels='Unnamed: 32', axis=1, inplace=True)
data.head()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
predictors = data.columns.drop('diagnosis')
target = "diagnosis"

train_data, test_data, train_sln, test_sln = train_test_split(data[predictors], data[target], test_size = 0.2, random_state=1)

#Normalize Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_data)
train_data_normalized = scaler.transform(train_data)
test_data_normalized = scaler.transform(test_data)

#now you can use train_data_normalized, test_data_normalized, train_sln, and test_sln
```

## Exercises:
1. Random Forest: \\
  a. Default \\
  b. Tuned

```python
#Default Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

rf = RandomForestClassifier(random_state=10)
rf.fit(train_data_normalized,train_sln)
predictions = rf.predict(test_data_normalized)
print("Default Random Forest Accuracy Score:",metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

```python
#Tuned Random Forest

tuned_rf = RandomForestClassifier(random_state=10, n_estimators=3)
tuned_rf.fit(train_data_normalized,train_sln)
predictions = tuned_rf.predict(test_data_normalized)
print("Tuned Random Forest Accuracy Score:",metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

2. Support Vector Classifier: \\
  a. Default \\
  b. Tuned

```python
#Default SVC
from sklearn.svm import SVC

svc = SVC(random_state=10)
svc.fit(train_data_normalized,train_sln)
predictions = svc.predict(test_data_normalized)
print('Default SVC Accuracy Score:',metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

```python
#Tuned SVC:

tuned_svc = SVC(random_state=10,C=2,tol=1)
tuned_svc.fit(train_data_normalized,train_sln)
predictions = tuned_svc.predict(test_data_normalized)
print('Tuned SVC Accuracy Score:',metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

3. Perceptron: \\
  a. Default \\
  b. Tuned

```python
#Default Perceptron
from sklearn.linear_model import Perceptron

perc = Perceptron(random_state=10)
perc.fit(train_data_normalized,train_sln)
predictions = perc.predict(test_data_normalized)
print('Default Perceptron Accuracy Score:',metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

```python
#Tuned Perceptron

tuned_perc = Perceptron(random_state=10,tol=1,eta0=17)
tuned_perc.fit(train_data_normalized,train_sln)
predictions = tuned_perc.predict(test_data_normalized)
print('Tuned Perceptron Accuracy Score:',metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

4. Stochastic Gradient Descent: \\
  a. Default \\
  b. Tuned

```python
#Default SGD
from sklearn.linear_model import SGDClassifier

SGD = SGDClassifier(random_state=10)
SGD.fit(train_data_normalized,train_sln)
predictions = SGD.predict(test_data_normalized)
print('Default SGD Accuracy Score:',metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

```python
#Tuned SGD

tuned_SGD = SGDClassifier(random_state=10,early_stopping=True,validation_fraction=0.4)
tuned_SGD.fit(train_data_normalized,train_sln)
predictions = tuned_SGD.predict(test_data_normalized)
print('Tuned SGD Accuracy Score:',metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```

5. Now take the model that performed the best above and run a **principal component analysis** on it. 

```python
#PCA on best model (tuned perceptron)
from sklearn.decomposition import PCA
extractor = PCA(n_components=2, whiten=True)
extractor.fit(train_data_normalized)
transform_train_data = extractor.transform(train_data_normalized)
transform_test_data = extractor.transform(test_data_normalized)

tuned_perc = Perceptron(random_state=10,tol=1,eta0=17)
tuned_perc.fit(transform_train_data,train_sln)
predictions = tuned_perc.predict(transform_test_data)
print('Tuned Perceptron Accuracy Score:',metrics.accuracy_score(test_sln,predictions))

vals = data[target].unique()
cm = metrics.confusion_matrix(test_sln, predictions, labels=vals)
print(pd.DataFrame(cm, index = "True " + vals, columns = "Predicted " + vals))
```
